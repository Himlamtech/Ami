````python
import pandas as pd

data_frame = pd.read_csv("data/processed/qa.csv")
data_frame
````

````python
processed_data = data_frame[["Câu Hỏi", "Câu Trả Lời"]]
````

````python
def create_context(data_frame):
    for i in range(len(data_frame)):
        data_frame.loc[i, "Context"] = (
            f"#Question : ```{data_frame.loc[i, 'Câu Hỏi']}```\n#Answer : ```{data_frame.loc[i, 'Câu Trả Lời']}```"
        )
    return data_frame


data_frame = create_context(data_frame)
data_frame.to_csv("data/processed/qa.csv", index=False)
````
